```python
# As we want to predict e-scooter movement we need to have future arrival data. 
# So we first have to come up with code that generates tomorrow’s date and transforms it into the format the endpoint 
# requires. To get there we use the datetime module by adding one day using datetime.timedelta()and format the string 
# according to the syntax requirements of the request. The AeroDataBox API only returns flight data for 12 hours. 
# Since we need data for the whole day we have to loop through two different time ranges and make a request for each of them: 
import pytz
from datetime import datetime, date, timedelta
import requests

tz = pytz.timezone('Europe/Berlin')
today = datetime.now().astimezone(tz).date()
times = [["00:00","11:59"],["12:00","23:59"]]
tomorrow = (today + timedelta(days=1))
icao = "EDDF"
tomorrow
```

```python
for time in times:
    url = f"https://aerodatabox.p.rapidapi.com/flights/airports/icao/{icao}/{tomorrow}T{time[0]}/{tomorrow}T{time[1]}"
    querystring = {
        "withLeg":"true",
        "direction":"Arrival",
        "withCancelled":"false",
        "withCodeshared":"true",
        "withCargo":"false",
        "withPrivate":"false"
        }
    headers = {
        'x-rapidapi-host': "aerodatabox.p.rapidapi.com",
        'x-rapidapi-key': "41ff370a44mshbe20330f78fd706p154d9ajsnaa91e9e7fba0" # put your API key here
        }
    response = requests.request("GET", url, headers=headers, params=querystring)
    flights_json = response.json()
```

```python
# For each of the two time frames we now have to loop through every flight in the JSON response to extract the data we need.
# So we’ll nest this loop inside the loop for the different time frames and fill the empty list flights_dict = {}
# by using the .get()method. 
# This makes sure that the code doesn’t break when reading non existing values or non existing keys: 
# if there is no text a None value will be inserted instead. For details see the code block below:
for time in times:
    url = f"https://aerodatabox.p.rapidapi.com/flights/airports/icao/{icao}/{tomorrow}T{time[0]}/{tomorrow}T{time[1]}"
    querystring = {
        "withLeg":"true",
        "direction":"Arrival",
        "withCancelled":"false",
        "withCodeshared":"true",
        "withCargo":"false",
        "withPrivate":"false"
        }
    headers = {
        'x-rapidapi-host': "aerodatabox.p.rapidapi.com",
        'x-rapidapi-key': "41ff370a44mshbe20330f78fd706p154d9ajsnaa91e9e7fba0" # put your API key here
        }
    response = requests.request("GET", url, headers=headers, params=querystring)
    flights_json = response.json()
    
    flights_dict = {}
    for flight in flights_json['arrivals']:
        flights_dict['arrival_time_local'] = flight['arrival'].get('scheduledTimeLocal', None)
        flights_dict['arrival_terminal'] = flight['arrival'].get('terminal', None)
        flights_dict['departure_city'] = flight['departure']['airport'].get('name', None)
        flights_dict['departure_icao'] = flight['departure']['airport'].get('icao', None)
        flights_dict['departure_time_local'] = flight['departure'].get('scheduledTimeLocal', None)
        flights_dict['airline'] = flight['airline'].get('name', None)
        flights_dict['flight_number'] = flight.get('number', None)
        flights_dict['data_retrieved_on'] = datetime.now().astimezone(tz).date()
```

```python
# Finally we wrap up the code in a function so we’re able to automatically request 
# the data for more than one airport via the icao code.
def tomorrows_flight_arrivals(icao_list):
    tz = pytz.timezone('Europe/Berlin')
    today = datetime.now().astimezone(tz).date()
    time = [["00:00","11:59"],["12:00","23:59"]]
    tomorrow = (today + timedelta(days=1))
    
    list_for_df = []

    for icao in icao_list:
        times = [["00:00","11:59"],["12:00","23:59"]]
    
        for time in times:
            url = f"https://aerodatabox.p.rapidapi.com/flights/airports/icao/{icao}/{tomorrow}T{time[0]}/{tomorrow}T{time[1]}"
            querystring = {
              "withLeg":"true",
              "direction":"Arrival",
              "withCancelled":"false",
              "withCodeshared":"true",
              "withCargo":"false",
              "withPrivate":"false"
                  }
            headers = {
              'x-rapidapi-host': "aerodatabox.p.rapidapi.com",
              'x-rapidapi-key': "41ff370a44mshbe20330f78fd706p154d9ajsnaa91e9e7fba0"
                  }
            response = requests.request("GET", url, headers=headers, params=querystring)
            flights_json = response.json() 
            flights_dict = {}  
      
            for flight in response.json()['arrivals']:
                flights_dict['arrival_icao'] = icao
                flights_dict['arrival_time_local'] = flight['arrival'].get('scheduledTimeLocal', None)
                flights_dict['arrival_terminal'] = flight['arrival'].get('terminal', None)
                flights_dict['departure_city'] = flight['departure']['airport'].get('name', None)
                flights_dict['departure_icao'] = flight['departure']['airport'].get('icao', None)
                flights_dict['departure_time_local'] = flight['departure'].get('scheduledTimeLocal', None)
                flights_dict['airline'] = flight['airline'].get('name', None)
                flights_dict['flight_number'] = flight.get('number', None)
                flights_dict['data_retrieved_on'] = datetime.now().astimezone(tz).date()
                list_for_df.append(flights_dict)
    return pd.DataFrame(list_for_df)
```
